```python
# Importing of libraries for later use.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Script to browse and upload csv file saved to local computer.
from google.colab import files

uploaded = files.upload()
```

```python
import io
```

```python
# Imported IO which is library to read csv into CoLab, printed df to see imported data.
df = pd.read_csv(io.BytesIO(uploaded['countryvaccinations.csv']))
print(df)
```

```python
# Created dataframe subsetting the Top 10 Countries with greatest values of total vaccinations
dfTop10 = df[:10]
print(dfTop10)
```

```python
# Plotted data from subset dataframe to stacked bar chart below. 
ax = dfTop10.plot.bar(stacked=True)
ax.set_xlabel('Country')
ax.set_ylabel('Vaccines in Billions')
ax.set_title('Vaccines by Country')
ax.legend(labels=['Total Vaccinations', 'Total People Partially Vaccinated', 'Total People Fully Vaccinated', 'Total Daily Vaccinations' ])
ax.set_xticklabels(['China', 'USA', 'India', 'UK', 'England', 'Brazil', 'Germany', 'France', 'Turkey', 'Italy'])
ax.tick_params(axis='x', labelrotation = 45)
```
